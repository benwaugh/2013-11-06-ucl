# Introduction

Aim is to plot dimuon mass spectrum from ROOT Tree in provided files events.root and events2.root. As is often the case, we are not given any formal specification and have to use some trial and error to work out what to do.

We can use ROOT to open an inspect the data, and plot distributions of variables. When we want to carry out some more complicated task such as finding pairs of opposite-charged particles and plotting their invariant mass, we need to start writing some code. We could do this using CINT, but this has a few disadvantages:

- the language understood by the interpreter is not quite the same as the C++ we can compile
- C++ is not an ideal language for quick prototyping and exploration: this is the main motivation for introducing the differences already referred to

A scripting language with dynamic typing is better for exploring problems, and the same code run from the command line can be saved as a script and even refactored into a more sophisticated application using classes and modules. Python is a popular choice in the sciences, and there is an API enabling ROOT classes and data formats to be used from Python.

# Getting started

Start ROOT, open a `TBrowser` and use it to see what variables are included in the data tree provided.

- `nPart` = number of particles in event
- `E` = energy of particles: from range of values can guess units are GeV
- `qpt` = signed transverse momentum [GeV]
- `phi` = azimuthal angle [radians]
- `theta` = polar angle [radians]

Before publishing our analysis we should make sure we *really* know what the data mean and how they were produced, what selection has already been applied and where defects could have crept in due to faulty code or other problems.

Let's start with a simple example:

```python
from ROOT import TCanvas 
c = TCanvas() 
```

We can get help on classes within ROOT, or instances:

```python
help(c)
```

Updating an existing `TCanvas` works OK in Python or IPython, but not in the IPython notebook, or at least I haven't managed to make it work. However, we can close the existing canvas by hand or programmatically ...

```python
c.Close()
```

... and display something else, e.g. a function, in its own automatically created canvas:

```python
from ROOT import TF1
f = TF1("sinc(x)", "sin(x)/x", 0, 10)
f.Draw()
```
